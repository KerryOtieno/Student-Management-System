## Question 1 : Student Result Management System

```python

# STUDENT RESULT MANAGEMENT SYSTEM

students = {}

def calculate_grade(score):
    if 70 <= score <= 100:
        return "A"
    elif 60 <= score <= 69:
        return "B"
    elif 50 <= score <= 59:
        return "C"
    elif 45 <= score <= 49:
        return "D"
    else:
        return "F"

def add_student():
    name = input("Enter student name: ")
    try:
        score = float(input("Enter score (0-100): "))
        if score < 0 or score > 100:
            print("Score must be between 0 and 100.")
            return
        students[name] = score
        print("Student added successfully.")
    except ValueError:
        print("Invalid score entered.")

def view_students():
    if not students:
        print("No student records found.")
        return

    print("\\nName\\tScore\\tGrade")
    for name, score in students.items():
        print(f"{name}\\t{score}\\t{calculate_grade(score)}")

def update_score():
    name = input("Enter student name to update: ")
    if name in students:
        try:
            score = float(input("Enter new score: "))
            if score < 0 or score > 100:
                print("Score must be between 0 and 100.")
                return
            students[name] = score
            print("Score updated successfully.")
        except ValueError:
            print("Invalid score entered.")
    else:
        print("Student not found.")

def search_student():
    name = input("Enter student name: ")
    if name in students:
        score = students[name]
        print(f"Name: {name}, Score: {score}, Grade: {calculate_grade(score)}")
    else:
        print("Student not found.")

def class_statistics():
    if not students:
        print("No records available.")
        return

    scores = list(students.values())
    print("Highest Score:", max(scores))
    print("Lowest Score:", min(scores))
    print("Average Score:", sum(scores)/len(scores))
    print("Total Students:", len(scores))

def student_menu():
    while True:
        print("\\n1. Add Student")
        print("2. View Students")
        print("3. Update Score")
        print("4. Search Student")
        print("5. Class Statistics")
        print("6. Exit")

        choice = input("Choose an option: ")

        if choice == "1":
            add_student()
        elif choice == "2":
            view_students()
        elif choice == "3":
            update_score()
        elif choice == "4":
            search_student()
        elif choice == "5":
            class_statistics()
        elif choice == "6":
            print("Exiting Student Result Management System.")
            break
        else:
            print("Invalid choice.")
```

## Question 2 : Inventory Management System

```python

# INVENTORY MANAGEMENT SYSTEM

inventory = {}

def add_product():
    product = input("Enter product name: ")
    try:
        qty = int(input("Enter quantity: "))
        if qty < 0:
            print("Quantity cannot be negative.")
            return
        inventory[product] = qty
        print("Product added successfully.")
    except ValueError:
        print("Invalid quantity entered.")

def view_inventory():
    if not inventory:
        print("Inventory is empty.")
        return

    print("\\nProduct\\tQuantity")
    for product, qty in inventory.items():
        print(f"{product}\\t{qty}")

def update_quantity():
    product = input("Enter product name: ")
    if product in inventory:
        try:
            qty = int(input("Enter new quantity: "))
            if qty < 0:
                print("Quantity cannot be negative.")
                return
            inventory[product] = qty
            print("Quantity updated successfully.")
        except ValueError:
            print("Invalid quantity entered.")
    else:
        print("Product not found.")

def search_product():
    product = input("Enter product name: ")
    if product in inventory:
        print(f"{product}: {inventory[product]}")
    else:
        print("Product not found.")

def remove_product():
    product = input("Enter product name: ")
    if product in inventory:
        del inventory[product]
        print("Product removed successfully.")
    else:
        print("Product not found.")

def low_stock_report():
    low_stock = {p:q for p,q in inventory.items() if q < 5}

    if not low_stock:
        print("No low-stock items.")
        return

    print("Low Stock Products:")
    for product, qty in low_stock.items():
        print(product, qty)

def inventory_statistics():
    if not inventory:
        print("Inventory is empty.")
        return

    quantities = list(inventory.values())

    print("Total Products:", len(inventory))
    print("Highest Quantity Product:", max(inventory, key=inventory.get))
    print("Lowest Quantity Product:", min(inventory, key=inventory.get))
    print("Total Stock Available:", sum(quantities))

def inventory_menu():
    while True:
        print("\\n1. Add Product")
        print("2. View Inventory")
        print("3. Update Quantity")
        print("4. Search Product")
        print("5. Remove Product")
        print("6. Low Stock Report")
        print("7. Inventory Statistics")
        print("8. Exit")

        choice = input("Choose an option: ")

        if choice == "1":
            add_product()
        elif choice == "2":
            view_inventory()
        elif choice == "3":
            update_quantity()
        elif choice == "4":
            search_product()
        elif choice == "5":
            remove_product()
        elif choice == "6":
            low_stock_report()
        elif choice == "7":
            inventory_statistics()
        elif choice == "8":
            print("Exiting Inventory Management System.")
            break
        else:
            print("Invalid choice.")
```
